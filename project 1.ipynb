***Calculate the percentage of delayed flights for both ALASKA and AM WEST airlines***

```python
(data.alaska_delayed.sum()/(data.alaska_on_time.sum() + data.alaska_delayed.sum())*100).round(decimals=2)
```

## Overall Percentage Results
> - ALASKA airlines has about 13.27% of it's flight arrivals delayed
> - AM WEST has about 10.89% of it's flight arrivals delayed

***Overall, AM WEST has a lower percentage of delayed flight arrivals. For the next comparison, we will look at each individual city***

***Calculate the percentage of delayed fights for both ALASKA and AM WEST airlines for each city***

```python
alaska_delay_percent = pd.Series(data.alaska_delayed/(data.alaska_on_time + data.alaska_delayed)*100).round(decimals=2)
alaska_delay_percent
```

```python
amwest_delay_percent = pd.Series(data.amwest_delayed/(data.amwest_on_time + data.amwest_delayed)*100).round(decimals=2)
amwest_delay_percent
```

***Display current dataframe***

```python
data['alaska_delay_percent'] = alaska_delay_percent
data['amwest_delay_percent'] = amwest_delay_percent
data
```

***Create and display visual plot of delayed percentages for each city for both ALASKA and AM WEST airlines***

```python
data.plot(x='city',y=['alaska_delay_percent','amwest_delay_percent'])
```

## Individuial Percentage Results & Final Conclusion

> - ALASKA in Phoenix has the lowest delayed arrival percentage
