## Wide Dataset #2 - California Cities
This is a dataset of 482 Calofornia cities. We will tidy the data and make downstream analyses.

```python
import numpy as np
import pandas as pd
import re
%matplotlib notebook
```

***Read and display raw california cities dataset***

```python
cities = pd.read_csv('https://raw.githubusercontent.com/jakevdp/PythonDataScienceHandbook/master/notebooks/data/california_cities.csv')
cities
```

***Drop useless attributes*** <br>
***Correct grammatical spacing errors in 'city' column***

```python
cities.rename({"Unnamed: 0":"useless"}, axis = "columns", inplace=True)
cities.drop(["useless"], axis = 1, inplace = True)

def spacing(city):
    space = re.findall('[A-Z][^A-Z]*', city)
    return ' '.join(space)

cities['city'] = cities['city'].apply(spacing)
```

***Calculate Top 5: highest population***

```python
cities[['city', 'population_total']].sort_values('population_total', ascending = False).head(5)
```

***Calculate Top 5: lowest population***

```python
cities[['city', 'population_total']].sort_values('population_total', ascending = True).head(5)
```

***Calculate Top 5: highest total area***

```python
cities[['city', 'area_total_sq_mi']].sort_values('area_total_sq_mi', ascending = False).head(5)
```

***Calculate Top 5: lowest total area***

```python
cities[['city', 'area_total_sq_mi']].sort_values('area_total_sq_mi', ascending = True).head(5)
```
