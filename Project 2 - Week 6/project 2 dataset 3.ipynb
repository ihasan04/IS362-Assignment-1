## Wide Dataset #3 - NYC Leading Causes of Death
This is a dataset of the leading causes of deaths in New York City. We will tidy the data and make downstream analyses.

```python
import numpy as np
import pandas as pd
%matplotlib notebook
```

***Read and display raw nyc leading causes of death dataset***

```python
deaths = pd.read_csv('https://raw.githubusercontent.com/pepimartinez/NYC-DeathRates/master/data/raw/New_York_City_Leading_Causes_of_Death.csv')
deaths
```

***Tidy data (replace "." with "NaN")*** <br>
***Drop useless data*** <br>
***Reassign string data to int***

```python
deaths = deaths.replace(to_replace = '.', value = np.nan)
deaths.isnull().sum()
deaths.Deaths = deaths.Deaths.dropna().astype(np.int64)
```

***Sort data by year and deaths***

```python
deaths.sort_values(['Year', 'Deaths'], ascending=[True,False])
```

***Filter data for deaths for White Non-Hispanic Females*** <br>
***Sort by year and deaths***

```python
deaths[(deaths['Sex'] == 'F') & (deaths['Race Ethnicity'] == 'White Non-Hispanic')].sort_values(by = ['Year' , 'Deaths'])
```

***Group leading causes of deaths with race ethnicity*** <br>
***Sort deaths by race ethnicity***

```python
deaths_by_race = deaths.groupby(['Race Ethnicity', 'Leading Cause'], as_index = False)['Deaths'].sum()
deaths_by_race.sort_values(['Race Ethnicity', 'Deaths'], ascending = [True,False])
```

***Calculate Top 5: most leading causes of deaths in 2014***

```python
deaths[deaths['Year'] == 2014].sort_values('Deaths', ascending = False).head(5)
```

***Calculate Top 5: least leading causes of deaths in 2007***

```python
deaths[deaths['Year'] == 2007].sort_values('Deaths', ascending = True).head(5)
```
