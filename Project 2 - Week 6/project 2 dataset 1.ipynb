## Wide Dataset #1 - IMDB Movie Samples
This is a subset of 100 records from IMDB movie dataset. We will tidy the data and make downstream analyses.

```python
import numpy as np
import pandas as pd
%matplotlib notebook
```

***Read and display raw movie sample dataset***

```python
imdb_data = pd.read_csv('https://raw.githubusercontent.com/erajabi/Python_examples/master/movie_sample_dataset.csv', encoding='utf-8')
imdb_data
```

***Drop useless attributes, handle text and numeric attributes*** <br>
***Fill up the empty values of two numeric columns: gross and budget with 0***

```python
imdb_data.drop(['color','language'], axis = 1, inplace = True)
imdb_data['director_name'].fillna('', inplace = True)
imdb_data['gross'].fillna(0, inplace = True)
imdb_data['budget'].fillna(0, inplace = True)
```

***Unify the values for other columns like country, director name, or movie title***

```python
imdb_data['country'] = imdb_data['country'].str.upper()
imdb_data['country'] = np.where(imdb_data['country'] == 'UNITED STATES','USA', imdb_data['country'])
imdb_data['director_name'] = np.where(imdb_data['director_name']=='N/A','', imdb_data['director_name'])
imdb_data['director_name'] = np.where(imdb_data['director_name']=='Nan','', imdb_data['director_name'])
imdb_data['director_name'] = np.where(imdb_data['director_name']=='Null','', imdb_data['director_name'])
imdb_data['movie_title'] = imdb_data['movie_title'].str.replace('Â', '')
```

***Handle the outliers where duration time is lass than 10 or more than 300*** <br>
***Filling the correspondent rows with appropriate numbers*** <br>
***0 value is used for those outliers***

```python
imdb_data["gross"] = imdb_data["gross"].astype(float)
imdb_data["duration"] = imdb_data["duration"].astype(float)
imdb_data["budget"] = imdb_data["budget"].astype(float)

imdb_data['duration'] = np.where(imdb_data['duration'] <= 10,0, imdb_data['duration'])
imdb_data['duration'] = np.where(imdb_data['duration'] > 300,0, imdb_data['duration'])
imdb_data['imdb_score'] = np.where(imdb_data['imdb_score'] <= 0,0, imdb_data['imdb_score'])
imdb_data['title_year'] = np.where(imdb_data['title_year'] < 2010,0, imdb_data['title_year'])
```

***Normalize data by splitting up actors into three different attributes and drop the actual column***

```python
actor_list = imdb_data["actors"].str.split(",", n = 2, expand = True) 
imdb_data["actor1"] = actor_list[0] 
imdb_data["actor2"] = actor_list[1] 
imdb_data["actor3"] = actor_list[2] 
imdb_data.drop(columns = ['actors'], inplace = True)
```

***Add a new metric called GOB that shows gross over budget for each movie***

```python
imdb_data['GOB'] = imdb_data.apply(lambda row: row['gross'] / row['budget'] if row['budget'] != 0 else 0, axis = 1)
```
